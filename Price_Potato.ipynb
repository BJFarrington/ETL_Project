```python
#import Dependencies
import pandas as pd
import csv
import matplotlib.pyplot as plt
import os
import sys
import numpy as np
```

```python
#Read in CSV file with Price data per month
csv_price = 'Consumer_Price_index_Potatoes10yrs.csv'
price_df = pd.read_csv(csv_price)
price_df.head()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

#Scatter Plot 2015
x_axis = price_df['Month-Yr']
y_axis = price_df['Price']
plt.scatter(x_axis, y_axis, marker="o", facecolors="red", edgecolors="black", alpha=0.8, label="Potato Price")

ax.set_xticks(ax.get_xticks()[::4])
plt.xticks(rotation=90)

#Add title to our scatter plot along with an x-axis and y-axis title
plt.title("CPI-Average Price Data")
plt.xlabel("Month-Year")
plt.ylabel("Price / per lb.")

plt.legend(loc='lower right')

plt.savefig('month_price_scatter.png')

plt.show()
```
