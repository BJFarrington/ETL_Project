```python
#import Dependencies
import pandas as pd
import csv
import matplotlib.pyplot as plt
import os
import sys
import numpy as np
```

```python
#Read in CSV file with Price data per month
csv_price = 'Consumer_Price_index_Potatoes10yrs.csv'
price_df = pd.read_csv(csv_price)
price_df.head(3)
```

```python
#Normalizing for potato price inflation
url = 'http://www.in2013dollars.com/Potatoes/price-inflation'

#extract all tables into a list
tables = pd.read_html(url)

#save 2nd table from list in to dataframe
potato_inflation_df = tables[1]

potato_inflation_df
```

```python
Price_inflation_df = pd.merge(price_df, potato_inflation_df, on='Year')
Price_inflation_df
```

```python
Price_inflation_df.to_csv('Price_inflation.csv')
```
