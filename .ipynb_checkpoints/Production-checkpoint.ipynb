```python
import pandas as pd
```

```python
potato_prod='Data/raw_data/potato_prod.csv'
potato_yield='Data/raw_data/potato_yield.csv'
```

```python
prod_df=pd.read_csv(potato_prod)
prod_df
```

```python
potato_prod = prod_df.drop(prod_df.index[[0,1,12,23,34,36]])
potato_prod.head()
```

```python
yield_df=pd.read_csv(potato_yield)
yield_df
```

```python
potato_yield = yield_df.drop(yield_df.index[[0,1,12,23,34,36]])
potato_yield.head()
```

```python
export_csv = potato_prod.to_csv(r'Data/clean_data/potato_prod.csv', index = None)
```

```python
export_csv = potato_yield.to_csv(r'Data/clean_data/potato_yield.csv', index = None)
```
