```python
#import Dependencies
import pandas as pd
import csv
import matplotlib.pyplot as plt
import os
import sys
import numpy as np
```

```python
#Read in CSV file with Price data per month
csv_price = 'Consumer_Price_index_Potatoes10yrs.csv'
price_df = pd.read_csv(csv_price)
price_df.head()
```

```python
# Seprate coumn"Month-Yr" to two columns"Year", "Month"
price_df["Month"] = price_df["Month-Yr"].apply(lambda x:x.split("-")[1])
                                               
price_df["Year"] = price_df["Month-Yr"].apply(lambda x:x.split("-")[0])
# Convert 1-dig or 2-dig to 4-dig for column year 
price_df["Year"] = price_df["Year"].apply(lambda x:pd.to_numeric("20" + ("00"+x)[-2:]))

price_df = price_df.drop(columns=["Month-Yr"])
```

```python
price_df.head()
```

```python
#Normalizing for potato price inflation 
url = 'http://www.in2013dollars.com/Potatoes/price-inflation'

#extract all tables into a list
tables = pd.read_html(url)

#save 2nd table from list in to dataframe
potato_inflation_df = tables[1]

potato_inflation_df.head()
```

```python
# We only keep data after 2009
potato_inflation_df=potato_inflation_df[potato_inflation_df["Year"]>2008].sort_values(by="Year")
potato_inflation_df
```

```python
Price_inflation_df = pd.merge(price_df, potato_inflation_df, on='Year')
Price_inflation_df.head()
```
